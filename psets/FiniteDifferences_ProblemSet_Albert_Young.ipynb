# Part I: Particle in a box with linear ramp

### 1.1 Define a function called PIB_1D_linear_ramp with takes as input an array $x$ and a parameter $b$ and which returns a variable $Vx = bx$

```python
def PIB_1D_linear_ramp(x, b):
    return x * b
```

### 1.3 Use the finite differences approach to obtain and plot the the ground state wave function and associated pdf for different values of b. <br> Also calculate the expectation value of x, $<x>$ for each value of b.                                                                            

For this, use the range of values b_array = [0.01, 0.1, 1, 10]. Write a for-loop such that, for each value of b in b_array, you do the following: 

<ol>
<li>Generate the potential Vx for that value of b</li>
<li>Generate the H matrix associated to that potential using quantumWorld.build_H_matrix</li>
<li>Get the full set of eigenvalues and eigenvectors of H</li>
<li>Get the index for the lowest energy eigenvalue (use numpy.argsort)</li>
<li>Get the eigenfunction associated to the lowest eigenvalue, calling it psi0</li>
<li>Normalize this eigenfunction using quantumWorld.normalize_wf, calling this psi0_norm</li>
<li>Get the pdf associated to this eigenfunction (call it pdf)</li>
<li>Get $<x>$ for this eigenfunction, and store in a list x_exp_array</li>
<li>In a single figure, plot, overlayed, the ground state eigenfunctions for each value of b.</li>
<li>In another figure, plot, overlayed, the pdf associated to the ground state for each value of b.</li>
</ol>

```python
%qtconsole
```

```python
L = 5
x = np.arange(0,L,0.01)
b_array = [0.01, 0.1, 1, 10]
x_exp_array = [] #empty array to store <x>
for b in b_array :
    #Get Vx, the matrix H, and its eigenvalues and eigenvectors.
    Vx = PIB_1D_linear_ramp(x, b)
    H = quantumWorld.build_H_matrix(x, Vx)
    eigenvalues, eigenvectors = np.linalg.eig(H)
    
    #Next, we sort the eigenvalues from least to largest energy, 
    #and get the indices of the sorted eigenvalues
    sort_index = numpy.argsort(eigenvalues)
    ind1 = sort_index[0] #This is the index of the lowest energy eigenvalue
    
    #Get the ground state eigenfunction
    psi0 = eigenvectors[:,ind1]
    #Normalize it
    psi0_norm = quantumWorld.normalize_wf(x, psi0)
    #Get the pdf associated to the ground state
    pdf = quantumWorld.probabilityDensity(psi0_norm)

    #Get the expectation value of x, <x>, call it x_exp
    integrand = pdf * x
    x_exp = simps(integrand, x)
    #append <x> to the list where we'll store <x>'s for each value of b
    x_exp_array.append(x_exp) 
    
    #plot the normalized ground state eigenfunctions, overlayed in single figure
    plt.figure(1)
    plt.xlabel('x')
    plt.ylabel('Psi(x)')
    plt.plot(x, psi0_norm, linewidth=2)
    
    #and the associated pdf's overlayed in another figure
    plt.figure(2)
    plt.xlabel('x')
    plt.ylabel('pdf')
    plt.plot(x, pdf, linewidth=2)
    
#Add legends to both figures:    
legend_str = ['b='+str(b) for b in b_array ]
plt.figure(1)
plt.legend(legend_str)
plt.figure(2)
plt.legend(legend_str)
```

### 1.4 Plot the expectation value of x as a function of the parameter b. Use quantumWorld.chem160_plotting() function

```python
quantumWorld.chem160_plotting([b_array], [x_exp_array], title='<x> vs. b', xlabel='b', ylabel='<x>')
```

# Part II: Diatomic molecule model with Coulomb potential

### 2.1 Define a function called coulomb_potential_1D_double which generates a potential consisiting of a double well, each well falling as $f(x)=-\frac{1}{(x-x_i)}$, where $x_i$ is the center of the i-th well.  The inputs to the function will be x, a numpy array defining the domain, and R, the separation between the two centers of the wells. 

I'll write this one for you...

```python
def coulomb_potential_1D_double(x,R):
    '''Create a double well, coulomb-like potential.
    Inputs:
    x --> numpy array of positions
    R --> the distance between the centers of the two wells
    '''
    Vx = np.zeros_like(x)
    L = max(x) - min(x) #length of domain
    x0_1 =  (L - R) / 2.0 #the first well
    x0_2 = x0_1 + R #center of the second well
    for i in range(len(Vx)):
        if x[i] == x0_1 or x[i] == x0_2:
            Vx[i] = Vx[i - 1]
        elif x[i] < max(x) / 2.0:
            Vx[i] = - 1.0 / np.abs(x[i] - x0_1)
        else:
            Vx[i] = - 1.0 / np.abs(x[i] - x0_2)
    return Vx
```

### 2.2   Create an x array that goes from $x=0$ to $x = 2$. And set R, the separation between the wells, to 0.5. Generate the potential, call it Vx, and plot it.

```python
x = np.arange(0, 2, 0.01)
R = 0.5
Vx = coulomb_potential_1D_double(x, R)

fig, ax = plt.subplots(1)
fig.set_size_inches(10, 8)
plt.plot(x, Vx, linewidth = 4)
#Some plot parameters
plt.xlabel('x -- position', fontsize = 16)
plt.ylabel('V(x)', fontsize = 16)
plt.title('Coulomb potential', fontsize = 16)
```

### 2.3 Build H Matrix (use the function quantumWorld.build_H_matrix) associated to this potential, and get its eigenvalues and eigenfunctions. 

```python
H = quantumWorld.build_H_matrix(x, Vx)
eigenvalues, eigenfunctions = np.linalg.eig(H)
```

### 2.4 Plot the two lowest energy eigenfunctions in one figure (overlayed) and their associated pdf's in another figure, overlayed

```python
#We'll use the numpy.argsort function to 
#sort the eigenvalues and get the indices of the 
#sorted eigenvalues
sort_index = numpy.argsort(eigenvalues)
ind1 = sort_index[0] #index of the lowest energy eigenvalue
ind2 = sort_index[1] #index of the second lowest one

#Get the ground state eigenfunction
psi0 = eigenfunctions[:,ind1]
psi1 = eigenfunctions[:,ind2]
#Normalize it
psi0_norm = quantumWorld.normalize_wf(x, psi0)
psi1_norm = quantumWorld.normalize_wf(x, psi1)

#Get the pdf associated to the ground state
pdf0 = quantumWorld.probabilityDensity(psi0_norm)
pdf1 = quantumWorld.probabilityDensity(psi1_norm)

#and we use these indices to plot the eigenfunctions associated to them

plt.figure(1)
plt.xlabel('x')
plt.ylabel('Psi(x)')
plt.plot(x, psi0_norm, linewidth=2, hold=True)
plt.plot(x, psi1_norm, linewidth=2)

#and the associated pdf's overlayed in another figure
plt.figure(2)
plt.xlabel('x')
plt.ylabel('pdf')
plt.plot(x, pdf0, linewidth=2, hold=True)
plt.plot(x, pdf1, linewidth=2)

#Add legends to both figures:    
legend_str = ['n=0','n=1']
plt.figure(1)
plt.legend(legend_str)
plt.figure(2)
plt.legend(legend_str)
```

### Plot the energy of the two lowest eigenvalues as a function of the separation R, and capture the splitting of the two lowest energy eigenstates as the wells approach each other. 

```python
x = np.arange(0, 2, 0.01)
R = np.arange(0.1, 2, 0.1)

E_1 = []
E_2 = []

for r in R:
    Vx = coulomb_potential_1D_double(x, r)
    H = quantumWorld.build_H_matrix(x, Vx)
    eigenvalues, eigenfunctions = np.linalg.eig(H)
    sort_index = numpy.argsort(eigenvalues)
    
    ind1 = sort_index[0] #index of the lowest energy eigenvalue
    ind2 = sort_index[1] #index of the second lowest one
    
    E_1.append(eigenvalues[ind1])
    E_2.append(eigenvalues[ind2])

quantumWorld.chem160_plotting([R, R], [E_1, E_2], title='Two lowest energies vs. R', legend_label=('E1', 'E2'), xlabel='R (separation)', ylabel='E')
```
