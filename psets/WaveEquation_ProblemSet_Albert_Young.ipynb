# Introduction


In this iPython notebook, we will explore solutions to the wave equation.  As you'll learn throughout the course, waves are fundamental to quantum mechanics, and playing around with the wave equation in the early part of the course will sharpen your intution later on down the road. 

In David's pencasts on the wave equation we saw that the analytical solution to the wave equation that describes a 1-D string can obtained 'separation of variables' technique.  The solution is given in general form by: 

$$ A(x,t)=\sum_n{c_n\cos(\frac{n\pi c}{L}t)\sin(\frac{n\pi x}{L})} $$

We'll refer to $ sin(\frac{n\pi x}{L}) $ as the **time independent portion**, and to $ cos(\frac{n\pi c}{L}t) $ as the **time dependent** portion. Recall that x is position, t is time.  In the case of a string, L is the length of the string and c is the velocity of sound on the string.  Finally, n is an integer that comes from satisfying the boundary conditions

That is, the wave can be expressed as a sum of "Eigenstates" or "Harmonics", each one characterized by a different number (an integer) n.

```python
#We'll import some useful modules for you. 
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from scipy.integrate import quad
```

**1a. Define a function that, given x, n, and L, returns the time-independent portion of an eigenstate. **

```python
def harmonic_time_independent(x,n,L):
    #### (ONE LINE)SUBSTITUTE THIS WITH A ONE-LINER THAT ASSIGNS THE 
    ####TIME-INDEPENDENT PORTION TO A VARIABLE CALLED amplitude
    amplitude = np.sin(n*np.pi*x/np.double(L))
    return amplitude
```

**1b. Define a function that, given t, c, n, and L, returns the time-dependent portion of an eigenstate.**

```python
def harmonic_time_dependent(t,c,n,L):
    #### (ONE LINE)SUBSTITUTE THIS WITH A ONE-LINER THAT ASSIGNS THE 
    ####TIME-DEPENDENT PORTION TO A VARIABLE CALLED amplitude
    amplitude = np.cos(n*np.pi*c*t/np.double(L))
    return amplitude
```

#Decomposition of wave into eigenstates

First, we create a function that returns a square wave: x = 0 if $x<L/3$ or $x>(2/3)L$.   x = 1 otherwise. 

```python
#We write the square function as an actual python function that 
#given an input x, returns a single output y
def square_function(x, L):
    '''given x, returns y as defined by the square function defined in the range 0 <= x <= L
    '''
    if x < 0 or x > L:
        print 'Error: the value of x should be between 0 and L'
        y = None
        return y
    elif x <= L/3.0 or x >= L*2.0/3:
        y = 0
    else:
        y = 1
    return y
```

Let's see what it looks like:

```python
L=10
x = np.arange(0,10,0.01)
square_y = []
for i in xrange(0, len(x)):
    square_y.append(square_function(x[i], L))
plt.plot(x,square_y)
plt.axis((0,L,0,1.5))
```

Now, follow what we did in the WaveDecompositionDemo.ipynb notebook in class. 

We want to show that this square-shaped wave can be expressed as a sum of sines, each one with a different frequency (different values of n):

$$ f(x) = \sum_1^{\infty}c_n\sin(\frac{n \pi x}{L})$$

Where the $c_n$ are the "AMPLITUDES" associated to each term.

In order to get these amplitudes, we need to "PROJECT" the triangle function into each of the of the sine functions.  This is done by taking the following inner product: 

$$ c_n = \frac{2}{L}\int_0^L \sin(\frac{n \pi x}{L})f(x)dx $$

#FOLLOW WHAT WE DID IN THE WaveDecompositionDemo.ipynb NOTEBOOK IN CLASS CLOSELY!!!
Write code below to get the amplitudes for the c1, c2, c3, ..., c10 coefficients.  Like we did in class you will need to<br>
1.  Write a projection_integrand function, that returns, for values of x, n and L, the product  $\frac{2}{L}\sin(\frac{n \pi x}{L})f(x)$, where f(x) is your square function evaluated at x <br>
2.  Write a foor loop that iterates over values of n, from 1 to 10 <br>
3.  for each value of n, use the quad function to evaluate the integral needed to get the coefficient c_n. <br>
4.  Add up all the $c_n\sin(\frac{n \pi x}{L})$, storing the in an array. <br>
5.  Plot the sum-of-sines approximation and the actual square wave function overlayed in the same figure. 

```python
def projection_integrand(x,n,L):
#     print(x, n, L)
    '''The inputs to the function are:
    x ---> vector of x values. 
    n ---> the n-number associated to the sine functions
    L --> L, upper limit of integration
    '''
    sine_function = np.sin(n*np.pi*x/np.double(L)) # this is the sine function sin(n*pi*x/L)
    integrand = (2.0/L)*sine_function*square_function(x, L) # this is the product of the two functions, with the 2/L factor
#     print(integrand)
    return integrand

n_max = 10
x = np.arange(0,L,0.01) # our x vector
square_approx = np.zeros(len(x))

for n in xrange(1, n_max + 1):
    c_n = quad(projection_integrand, 0, L, (n, L))
    
    #2.  update our square approximation, by adding up the c_n*sin(n*pi*x/L)
    square_approx = square_approx + c_n[0]*np.sin(n*np.pi*x/np.double(L))

    plt.hold(True)
plt.plot(x, square_approx)
plt.plot(x, square_y) 
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('approximating the square function as a sum of sines, n = 1 ...' + str(n_max))
plt.legend(['approximation', 'square function'])
```
