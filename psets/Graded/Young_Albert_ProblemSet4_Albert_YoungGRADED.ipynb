### 1.1 Now we'll set up a figure with 9 subplots (3 rows and 3 columns), and in each subplot we'll plot as a colormap one of the eigenstates

```python
#Set up 3-row, 3-column figure (9 subplots) 
fig, axes = plt.subplots(nrows=3, ncols=3)
fig.set_size_inches(14, 10) #of whatever size we want. 

#####################################
#Create a for loop that goes from 0 to len(l_m_list)
for i in xrange(len(l_m_list)):

    ####ALL THIS IS INSIDE THE FOR LOOP
    ax = axes.flat[i] #specify that we want to plot in the i-th subplot. 
    #############################################
    #get the i-th l and m numbers in l_m_list.  assign them to l and m, respectively. 
    l = l_m_list[i][0]
    m = l_m_list[i][1]

    ##############################################
    
    ##############################################
    #assign to the variable Psi_xy the (l,m) eigenfunction of the 2D HO. Use harmonic_oscillator_2D in 
    #your quantumWorld module. 
    Psi_xy = quantumWorld.harmonic_oscillator_2D(xx, yy, l, m)
    
    ##############################################
    
    #Plot as a color map. 
    p = ax.pcolor(x, y, Psi_xy)
    ax.axis((-3,3,-3,3))
    ax.set_title('l='+str(l)+', y='+str(m))   
    
#And the code below is just to set a common x and y label to the set of 9 figures.     
fig.text(0.5, 0.04, 'x coordinate', ha='center', va='center', fontsize = 14)
fig.text(0.06, 0.5, 'y coordinate', ha='center', va='center', rotation='vertical', fontsize = 14)
```

### 1.2 Now we will generate an initial linear combination of two eigenstates

```python
#We will work in atomic units
hbar = 1
omega = 1

#The l and m values of our two eigenstates which we will linearly combine. 
#Quantum numbers for the first eigenfunction we'll use
l1=0
m1=0
#Quantum numbers for the second eigenfunction we'll use. 
l2=1
m2=1

################
#Since we want a normalized linear combination, we will set the coefficients to 1/sqrt(2)
#Create variables c1_0 and c2_0 such that the linear combination of two 
#eigenfunctions is normalized. 
c1_0 = 1/np.sqrt(2)
c2_0 = 1/np.sqrt(2)
#####################


######################
#Let's call harmonic_oscillator_2D to get our two eigenstates (0,0) and (1,1), calling them 
#psi_xy_1 and psi_xy_2, respectively
psi_xy_1 = quantumWorld.harmonic_oscillator_2D(xx, yy, l1, m1, omega=omega, hbar=hbar)
psi_xy_2 = quantumWorld.harmonic_oscillator_2D(xx, yy, l2, m2, omega=omega, hbar=hbar)
#################

#######################
#Let's assign to E1 and E2 the energies of these two eigenstates. 
#Make sure you assign the right energy values!!!!
#What is the energy of the l=0, m=0 and the l=1 and m=1 eigenstates?
E1 = hbar*omega*(l1+m1+1)
E2 = hbar*omega*(l2+m2+1)
########################


#########################################
#Let's create the linear combination we want of the two eigenstates.
#calling it psi0
psi0 = c1_0 * psi_xy_1 + c2_0 * psi_xy_2
#####################################
```

### 1.3 Now, we will generate a time array, and for each time step, get the time-evolved values of c1 and c2, and plot the wavefunction as a color map. 

```python
######################
#Create a time array that goes from t=0 to t=3 in steps of 0.1
t_array = np.arange(0, 3, 0.1)

#############################

########################
#Write a for loop that loops over each value of the time array
for t in t_array:
    ##ALL THIS IS INSIDE THE FOR LOOP. 
    
    ###################
    #use the cn_t_function in quantumWorld to get the value of the cn coefficient at time = t. 
    #assign to variables c1_t and c2_t
    c1_t = quantumWorld.cn_t_function(c1_0, t, E1)
    c2_t = quantumWorld.cn_t_function(c2_0, t, E2)
    ###################
    
    ##################
    #assign to variable psi_t the updated linear combination of the two eigenfunctions
    psi_t = c1_t * psi_xy_1 + c2_t * psi_xy_2
    ##################
    
    #plot as a 2D color map. 
    #This is still inside the for loop. 
    fig, ax = plt.subplots()
    p = ax.pcolor(x, y, psi_t.real)
    ax.axis((-3,3,-3,3))
    cb = fig.colorbar(p, ax=ax)
```

# 2.  Propagating a Gaussian wave packet in a 1D harmonic oscillator

### In this section we will create an animation for the time-evolution of a Gaussian wave packet in a 1D harmonic oscillator.  In order to time evolve the wave packet, we will first need to compute the overlap integrals to get the different coefficients at $t=0$: $$ c_n(t=0) = \int \psi_n(x)\;\psi(x,t=0)\;dx$$, where the $\psi_n(x)$ are the eigenstates of the 1D Harmonic Oscillator, and $\psi(x)$ is our Gaussian wave packet.

### 2.0 First, let's go ahead and create and plot an initial Gaussian Wave Packet

```python
#####################
#First create an array of positions, x, which goes from -10 to 10 in intervals of 0.01

x = np.arange(-10, 10, 0.01)
#######################


#######################
#Let's assign to Vx the harmonic oscillator potential, using quantumWorld.harmonic_oscillator_v
Vx = quantumWorld.harmonic_oscillator_V(x)
###################


#We'll shift the gaussian wave packet by x0 to the right of the potential
x0 = 3
#Now, we'll assign to psi_0 a simple Gaussian wave packet.
psi_0 = np.exp(-(x-x0)**2)

#########################
#And normalize it using normalize_wf inside your quantumWorld module
#You can call the normalized wavefunction psi_0 again
psi_0 = quantumWorld.normalize_wf(x, psi_0)
#######################

#And now let's plot the gaussian wave packet, at t=0, overlayed with the potential. 
#we'll use a scaling factor just for visualization purposes. 
scaling_factor = 30
quantumWorld.chem160_plotting([x,x], [scaling_factor*psi_0, Vx], '1D-HO potential and Gaussian Wave Packet at t=0', None,'x - position', 'V(x) and Psi(x)' )


```

### 2.1 Now, let's get the $c_n$ coefficients by computing the overlap integrals $$ \int \psi_n(x)\;\psi(x,t=0)\;dx$$  where $\psi(x,t=0)$ is the initial Gaussian wave packet.

##### If you are lost as to what we are doing, you might want to review the demo WaveDecompositionDemo.ipynb as well as the problem set on waves, WaveEquationProblemSet_Filled.ipynb, where we approximate the triangle and square functions as sums of, in that case, sine functions. 

```python
######################
#How many eigenfunctions do we want to project agains?
#Set the variable nmax to however many you want to try first. 
#Don't go crazy... start with a few, say 5 or 10 or 15...
nmax = 30
#####################

##################
#Initialize an array of zeros, called psi_0_approx, which is the same length as psi_0
psi_0_approx = np.zeros_like(psi_0)

###################

###################
#Initialize an array of zeros, called c_n_0_array which is of length nmax
c_n_0_array = np.zeros(nmax)
###################

###########################
#Start a for-loop that loops over each value from n = 0 .... nmax
for n in xrange(nmax):
    
    ##ALL THIS IS INSIDE A FOR LOOP 
    #####################
    #For each value of n, get the nth eigenfunction of the 1D HO using harmonic_oscillator_wf in 
    #your quantumWorld library. assign it to the variable psi_n
    psi_n = quantumWorld.harmonic_oscillator_wf(x, n)
    #######################
    
    #########################
    #Assign to the variable integrand the integrand of the overlap integral
    integrand = psi_n * psi_0
    ###########################
    
    ###########################
    #integrate the integrand using simps to get the c_n_0 coefficient (the variable name we'll)
    #use for the c_n coefficient at time = 0. 
    c_n_0 = simps(integrand, x)
    ###########################
    
    ##########################
    #update psi_0_approx by adding the new term of the sum, c_n_0*psi_n
    psi_0_approx = psi_0_approx + c_n_0*psi_n
    ###########################
    
    #########################
    #Finally, add c_n_0 the the nth position of the c_n_0_array
    c_n_0_array[n] = c_n_0
    ##########################
```

### Now we'll overlay in the same plot, the gaussian as well as the sum of eigenstates of the HO that we just built that should approximate the gaussian.  Do they look similar? (They might look so similar, that you can't tell they're two lines!)

```python
quantumWorld.chem160_plotting([x,x], [psi_0, psi_0_approx], 'Gaussian and sum of eigenstates', ('Gaussian', 'Gaussian Approx'), 'position', 'Wavefunction')
```

### In order to quantify how similar the approximation is to the actual Gaussian, we'll compute the overlap between the two, $$overlap=\int \psi(x,t=0)\;\psi_{approx}(x,t=0)\;dx\;$$and call the error $1 - overlap$ <br><br>  See what the error is, and modify $nmax$ in the cell above (or two cells above) until the error is less than $10^{-10}$ <br>(HINT: you don't need more than 40 eigenfunctions to make the error small enough. )

```python
##########################################
#assign to the variable overlap the integral mentioned above. use simps to integrate. 
overlap = simps(psi_0 * psi_0_approx, x)
#############################

error = 1 - overlap
print 'the error is ' + str(error) + ' when I use ' + str(nmax) + ' HO eigenstates to approximate the Gaussian '
```

### Now comes the fun stuff.  We'll generate an animation that shows how the Gaussian wave packet evolves with time. 

```python
#import the animation module
from matplotlib import animation

#We'll set up the figure for you
fig = plt.figure()
ax = plt.axes(xlim=(-8, 8), ylim=(-1, 1)) #create single axis in figure
plt.xlabel('NONE')
plt.ylabel('NONE')
line, = ax.plot([], [], lw=2) #line is a plot.... 

#This init function will be called to create 
#the base frame upon which the animation takes place.
#we'll set it up for you as well
def init():
    line.set_data([], []) #sets the line data to nothing
    return line,  #return the line object

### COMPLETE THIS ANIMATION FUNCTION, WHICH WILL BE CALLED 
### ITERATIVELY FOR MANY VALUES OF T
def animate(t): 
    #Add a time scaling factor, to make the animation slower
    time_scaling = 0.3
    t = time_scaling*t
    
    #####################################
    #set to psi_t an array of zeros, same length as psi_0
    psi_t = np.zeros_like(psi_0)
    ########################################
    
    ###############################
    #start a for-loop that let's you grab each coefficient in c_n_0_array
    for n in range(len(c_n_0_array)):
        ############################
        #assign to c_n_0 the nth element in c_n_0_array
        c_n_0 = c_n_0_array[n]
        ####################################
        
        ######################################
        #Assign to En the energy of the nth eigenfunction of the 1D HO. 
        En = hbar * omega * (n + 0.5)
        ####################################
        
        #################################
        #use cn_t_function in quantumWorld library to get c_n_t, the value of the nth coefficient
        #at time t
        c_n_t = quantumWorld.cn_t_function(c_n_0, t, En)
        ############################################
        
        #########################################
        #assign to psi_n the nth eigenfunction of the 1D HO
        psi_n = quantumWorld.harmonic_oscillator_wf(x, n)
        ###############################
        
        ##########################################
        #update psi_t by adding c_n_t*psi_n
        psi_t = psi_t + c_n_t * psi_n
        #####################################
        
    ###################
    #assign to the variable pdf the pdf associated to psi_t
    pdf = quantumWorld.probabilityDensity(psi_t)
    ###################
    
    line.set_data(x, pdf)  #This is what you plot at each frame in the animation. 
    return line,

#Let's create the animation!!!
anim = animation.FuncAnimation(fig, animate, init_func=init, frames=200)
#save the animation
anim.save('GaussianWavePacket_PDF.mp4', 'mencoder' ,fps=10)
```

# 3. The vibrational selection rules

## In this section, we'll take the expectation value of the dipole operator for different eigenstates of the Harmonic Oscillator, and we'll discover the vibrational selection rules. 

##### In essence we want to evaluate the **transition moment integral**: $$ \int \psi_m^* \; \hat{\mu} \; \psi_n \;dx$$ where $\psi_n$ and $\psi_m$ are the two states involved in the transition, and   $\hat{\mu}$ is the dipole operator.<br> In general, the allowed transitions can be determined by symmetry considerations (e.g. integrals of **odd functions** are zero when integrated over all space), but for this course we'll determine them by brute force numerically. <br> <br> We will consider the case where $\hat{\mu} = x$, i.e. the case where the transition moment operator is the **electric dipole operator**. 

### Define a function transition_moment_integrand which takes as input x, and two quantum numbers n1, n2, and which returns the integrand $\psi_m^* \; \hat{\mu} \; \psi_n\;= \psi_m^* \; x \; \psi_n\;$ where $\psi_n$ is the nth eigenstate of the 1D Harmonic Oscillator

```python
################
##define the function described above here, all on your own!!!
#It should take you around 4 or 5 lines. 
def transition_moment_integrand(x, n1, n2):
    return np.conjugate(quantumWorld.harmonic_oscillator_wf(x, n1))*x*quantumWorld.harmonic_oscillator_wf(x, n2)
################
```

### As a first step, we will try to get a feeling for the pairs of quantum numbers $n$ and $m$ for which the integral $ \int \psi_m^* \; \hat{\mu} \; \psi_n \;dx$ will dissapear. 

### First plot the integrand $\psi_n^* \; \hat{\mu} \; \psi_n  = \psi_n^* \; x \; \psi_n $ (i.e. for the case $n=m$ ) for $n=0...8$

```python
#Set up 3-row, 3-column figure (9 subplots) 
fig, axes = plt.subplots(nrows=3, ncols=3)
fig.set_size_inches(14, 10) #of whatever size we want. 

#For each sublot. 
for i in range(len(axes.flat)): # this is one way to go over each subplot. 
    ax = axes.flat[i]
    n=i #We'll plot n=0 in the first subplot, n=1 in the second, etc. 
    
    ###############################################
    #Assign to the variable integrand the integrand of the transition moment integral
    # using the function we defined above. do it for the case m=n 
    integrand = transition_moment_integrand(x, n, n)
    #####################################################
    
    #plot the integrand. 
    ax.plot(x, integrand, linewidth = 4)
    ax.set_title('n='+str(n),fontsize=14)

#And the code below is just to set a common x and y label to the set of 9 figures.     
fig.text(0.5, 0.04, 'x coordinate', ha='center', va='center', fontsize = 18)
fig.text(0.06, 0.5, 'transition moment integrand', ha='center', va='center', rotation='vertical', fontsize = 18)
```

### Finally, we'll evaluate the transition moment integral for all possible pairs of eigenfunctions in the range $n = 0....10$.  We'll store the value of each integral in a matrix, and display the matrix using the matshow function. To do this, you will need to write a nested for loop (a for loop inside another for loop) in order to evaluate the transition moment integral for all pairs of quantum numbers in the range $n = 0 ... 9$

```python
#This is our array of positions.
x = np.arange(-10, 10, 0.01)

######################################
#Create a matrix of zeros that has dimensions 10 by 10. 
#Call it trans_moment
trans_moment = np.zeros(shape=(10,10))
##########################


################################################
#Create a nested for loop that goes over each value n1 = 0....9 and n2 = 0...9
for n1 in xrange(9):
    for n2 in xrange(9):
        ###ALL OF THIS IS INSIDE TWO (2) FOR LOOPS
        
        ####################################
        #And for each pair of values, evaluates the transition moment integrand
        #assign to variable integrand
        integrand = transition_moment_integrand(x, n1, n2)
        #################################
        
        ############################
        #And assigns the integral of that integrand (with simps) to the n1,n2 
        #element of the trans_moment matrix
        trans_moment[n1][n2] = simps(integrand, x)
        ###############################
        
```

```python
fig, ax = plt.subplots()
cax = ax.imshow(np.abs(trans_moment), interpolation='nearest')

#plt.imshow(np.abs(trans_moment))
ax.set_xlabel('m', fontsize = 16)
ax.set_ylabel('n', fontsize = 16)
ax.set_title('Transition Moment Integrals for pairs of quantum numbers m, n', fontsize = 15)

cbar = fig.colorbar(cax, ticks=[-1, 0, 1])

```

### How do you interpret the plot above in terms of Vibrational Selection Rules?

The transition moment integral evaluates to a nonzero quantity only if n1 and n2 differ by 1. Else zero. Higher values of n correspond to higher values of the transition moment integral.
