## 1.2 Construct a wave function psi_x which is a linear combination of the first four eigenstates of the 1D particle in a box. Use a box of length 10. Use the function box_1D_eigenfunction inside QuantumWorldLib (now imported as quantumWorld). 

$\psi(x) = \psi _1(x) + \psi _2(x)+ \psi _3(x)+ \psi _4(x)$ where $\psi _n(x) = \sqrt{\frac{2}{L}}\sin(\frac{n \pi x }{L})$'

## 1.4 Notice that the wave function you just created is not normalized. Write a function that takes as input a wave function and returns the same wave function but normalized. 

If a wavefunction is not properly normalized, you would need to do the following: <br><br>
if $ \int_0^L \psi^*(x)\psi(x)dx = A$ <br><br>Then dividing the wavefunction by $\sqrt{A}$ will lead to proper normalization:<br><br> $$\psi_{norm}(x)=\frac{\psi(x)}{\sqrt{A}}$$

```python
#Define a function called normalize_wf, which takes as input, x and psi_x, and returns the
#normalized wave function, wf_norm
def normalize_wf(x, psi_x):
    '''this function normalizes a wave function
    Input --> x, numpy array of position vectors
              psi_x, numpy array representing wave function, same length as x
    Output: 
            wf_norm --> normalized wave function
    ''' 
    #########(THREE LINES)
    # 1. Get the pdf associated to psi_x, assign to pdf
    # 2. Integrate the pdf over the entire range x.  Use simps and assign to integral_norm
    # 3. normalize the wavefunction by dividing psi_x by the square root of integral norm.  
    #Assign to wf_norm
    ############
    pdf = quantumWorld.probabilityDensity(psi_x)
    integral_norm = simps(pdf, x)
    wf_norm = psi_x / np.sqrt(integral_norm)
    return wf_norm
```

## 1.5 Use the function you created in 1.4 above to normalize the $\psi(x)$ you built in 1.2.  Double check that the new wave function is actually normalized by performing the appropriate integral

```python
#Assign to variable pdf_before the pdf associated to psi_x (before normalization)
pdf_before = quantumWorld.probabilityDensity(psi_x)
##########(ONE LINE)
#Perform the integral of pdf_before over the entire range x, using the simps function 
#assign this value to integral_before
#############
integral_before = simps(pdf_before, x)

#####(THREE LINES)
# 1. normalize the wave function psi_x, using the function normalize_wf you just made,
#assing to psi_norm
# 2. Assign to pdf_after, the pdf associated to the normalized wave function
# 3. Perform the integral of pdf_after over the entire range x, using the simps function 
#assing this value to integral_after
############
psi_norm = normalize_wf(x, psi_x)
pdf_after = quantumWorld.probabilityDensity(psi_norm)
integral_after = simps(pdf_after, x)

#print both values, integral_before and integral_after.  Is your new WaveFunction normalized?
print 'the value before normalization is --> ', integral_before
print 'the value after normalization is --> ', integral_after

#plot to check if wavefunction/pdf are correctly normalized
quantumWorld.chem160_plotting([x,x], [psi_norm,pdf_after], title='Psi_norm and PDF', legend_label=('psi_norm', 'pdf'), xlabel='x (position)', ylabel='psi_norm and pdf')
```

## 2.1 Write a function that takes as input an initial value of a coefficient, $c_n(0)$, a value of time, t, and the energy of the eigenstate associated to n, $E_n$, and returns $c_n(t)$

#Recall that the expression for the evolution of the coefficient $c_n$ is given by: <br>

#$$c_n(t) = c_n(0) e^{ \frac{-iE_n t}{\hbar} }$$

```python
#create a function called cn_t_function, which takes as input 
#cn_0 --> the initial value of the coefficient
#t --> a value of time
#E_n --> the energy eigenvalue associated to the nth eigenfunction of the particle in a box
#and returns as output cn_t, the value of the coefficient at time t
def cn_t_function(cn_0, t, E_n, hbar = 1):
    '''this function evolves the coefficient cn
    '''
    #write the exponent (recall that in python, 1j means sqrt(-1) (i.e. 'i'))
    exponent = -1j*E_n*t / hbar 
    
    ########(ONE LINE)
    #multiply e^exponent (use np.exp function for the exponential)
    #by cn_0, assign the product to the variable cn_t
    ###########
    cn_t = cn_0*np.exp(exponent)
    
    return cn_t
```

```python
#Set the values of L and the x array
L = 10
x = np.arange(0,L,0.01)

#Set the values of n1 and n2
n1 = 1
n2 = 2

#Set the initial values of the c_n coefficients
c1_0 = (1.0/np.sqrt(2)) #The coefficients c1 and c2 at time = 0 
c2_0 = (1.0/np.sqrt(2))

#####(TWO LINES)
#Set the values of E1 and E2 to the analytical energies of the n=1 and n=2 eigenstates 
#Recall that we have a function in our QuantumWorldLib module that does it
#it is called analytical_E_n_1D_PIB
############
E1 = quantumWorld.analytical_E_n_1D_PIB(n1, L)
E2 = quantumWorld.analytical_E_n_1D_PIB(n2, L)

######(TWO LINES)
#Assign to psi1_x and psi2_x the n=1 and n=2 eigenfunctions, respectively
#use box_1D_eigenfunction in our QuantumWorldLib module
###########
psi1_x = quantumWorld.box_1D_eigenfunction(x, L, n1)
psi2_x = quantumWorld.box_1D_eigenfunction(x, L, n2)

#Perform the linear combination, and assign it to psi0
psi0 = c1_0*psi1_x + c2_0*psi2_x #The wave function at time = 0

#Plot psi0, our wave function at time = 0
#plt.plot(x, psi0)
quantumWorld.chem160_plotting([x], [psi0], title='Psi at time = 0', legend_label=None, xlabel='x (position)', ylabel='Psi(0)')
```

In this exercise, you will work with the operator that is associated with the momentum and position observables. <br>

Then you will evaluate the expectation values of position and momentum as a function of time for the wave function that we constructed above.

## 3.1 The operator associated to the momentum observable is given by $\hat{p} = -i \hbar \frac{d}{dx}$. Write a function that, given as input a wave function $\psi(x)$ (that is, and array of values for the wave function), returns $\hat{p}\psi(x) =  -i \hbar \frac{d\psi(x)}{dx}$.  In order to take the derivative, we recommend using the numpy.gradient function, and which uses the finite diference method to take a first derivative. 

```python
#Create a function called momentum_operator that takes as input psi_x, 
#and returns -i*hbar*dpsi/dx
def momentum_operator(psi_x, dx, hbar = 1):
    '''this function applies the momentum operator to psi_x
    input : 
        --> psi_x : a numpy array, representing a wave function
        --> hbar: planck's constant / 2pi (defaults to atomic units)
    '''
    #assign to prefactor -i*hbar (recall that, in python, i is 1j) 
    prefactor = -1j * hbar
    
    ########### (ONE LINE)
    #assign to derivative the derivative of psi_x with respect to x.  Use the np.gradient function. 
    ############
    derivative = np.gradient(psi_x, dx)
    
    #return prefactor*derivative
    return prefactor * derivative
```

## 3.2 Now, we'll evolve the wave function we created above, but this time we'll evaluate at each time point the expectation value of position and of momentum. 

```python
#Create two empty lists, one which will hold the expectation of p for different times
p_array = []
#the other the expectation value of x, for different times. 
x_array = []

##########(ONE LINE)
#Create an array of times, from 0 to 100, in intervals of 1
###########
t_array = np.arange(0, 100, 1)

#Write a for loop, for each value of t in t_array:
for t in t_array:
    
    #######(TWO LINES)
    #assign to c1_t the value of c1 at time=t.  Use the function cn_t_function you created above.  
    #assign to c2_t the value of c2 at time=t.  Use the function cn_t_function you created above. 
    ###########
    c1_t = cn_t_function(c1_0, t, E1)
    c2_t = cn_t_function(c2_0, t, E2)
    
    #Assign to psi_x that linear combination c1(t)*psi_1 + c2(t)*psi_2
    psi_x = c1_t*psi1_x + c2_t*psi2_x
    
    
    #########(ONE LINE)
    #Using the momentum_operator function, act on psi_x with p, and assign to output 
    #to p_on_psi
    ###########
    dx = 0.01
    p_on_psi = momentum_operator(psi_x, dx)
    
    #Assign to p_integrand the product psi-star * p_on_psi
    p_integrand = np.conjugate(psi_x)*p_on_psi
    
    ###########(ONE LINE)
    #Using the simps function, integrate the integrand to get the expectation value of p. 
    #assign this to p_expectation
    ############
    p_expectation = simps(p_integrand, x)
    
    #Get the integrand psi-star * x * psi.  Assign it to x_integrand
    x_integrand = np.conjugate(psi_x)*x*psi_x
    
    ###########(ONE LINE)
    #Using simps, integrate it to get the expectation value of x. Assign to x_expectation. 
    ############
    x_expectation = simps(x_integrand)
    '''you also need to give x as input. 
    '''
    
    #Append <x>(t) and <p>(t) to the appropriate arrays. 
    p_array.append(p_expectation)
    x_array.append(x_expectation)


#Plot, in two supblots, both <p> and <x> as a function of time    
fig, ax = plt.subplots(2,1)
fig.set_size_inches(10, 8)
plt.subplot(2,1,1)
plt.plot(t_array, p_array, 'g-', linewidth = 4)
plt.xlabel('time', fontsize = 18)
plt.ylabel('<p>', fontsize = 24)
plt.title('Expection values of p and x vs. time', fontsize = 24)
plt.subplot(2,1,2)
plt.plot(t_array, x_array, 'r-', linewidth = 4)
plt.xlabel('time', fontsize = 18)
plt.ylabel('<x>',fontsize = 24)
```
