# 1 Ammonia Maser

The potential energy of the Ammonia molecule can be modeled by: 

$$V(x) = \frac{1}{2}k(x-x_0)^2 + be^{-c(x-x_0)^2}$$

Write a function called ammonia_maser that takes as input an x position array as well as the parameters k, x0, b, and c, and returns the ammonia molecule potential, Vx

```python
###################################
def ammonia_maser(x, k, x0, b, c):
    return 0.5*k*np.square(x-x0) + b*exp(-c*np.square(x-x0))
###########################################
```

### Next, we give you the parameters for the ammonia potential. Use them to: <br> 1.  Generate the potential Vx over the range of x values that we give you. <br> 2.  Create a H matrix using the finite differences code in the library (build_H_matrix). **make sure you use m_ammonia for the mass!!!!**<br> 3.  Get the eigenvalues and eigenvectors of the matrix. 

```python
k=0.07598
x0=5
b=0.05684
c=1.3696

m_ammonia = 4668 # atomic units

x = np.arange(0,10,0.01)

##################################################
#In three lines, generate potential Vx, build H matrix, and get 
#its eigenvectors and eigenvalues
Vx = ammonia_maser(x, k, x0, b, c)
H = quantumWorld.build_H_matrix(x, Vx, m=m_ammonia)
E, psi = np.linalg.eig(H)
sort_index = numpy.argsort(E)
######################################################
```

# 2. Transition probability of a wavepacket through a potential barrier as a function of particle energy and mass.

## Throughout this section, we'll refer to the beautiful diagram I made below, where an incoming free particle (from the right) interacts with a potential barrier.  A portion of the wave is reflected, and another portion is transmitted. 

![Potential Barrier](BarrierDiagram1.jpg)

For this section, we'll follow Schrier's Text book, Chapter 3, in our approach to simulate tunneling of a wavefunction through a potential barrier. To solve the tunneling problem with the finite differences approach we'll start out by rewriting the time-independent Schrodinger Equation: <br><br>
$$\hat{H}\psi = E\psi$$<br>
$$-\frac{\hbar^2}{2m}\frac{d^2\psi}{dx^2} + V(x)\psi = E\psi$$<br>
$$\frac{d^2\psi}{dx^2} + \frac{2m}{\hbar^2}[E - V(x)]\;\psi=0$$
<br>

Following the notation we used previously in the context of Finite Differences, $\psi_j$ will refer to the value of the wavefunction at the point $x_j$, that is $\psi(x_j)$.  Using the same expression for central-differences second derivative, we'll obtain: <br> 

$$\frac{\psi_{j+1} - 2\psi_j + \psi_{j-1}}{a^2} + \frac{2m}{\hbar^2}\;[E - V_j]\psi_j = 0 $$

<br>
Finally, we can rearrange this equation to write an expression for $\psi_{j+1}$ in terms of $\psi_j$ and $\psi_{j-1}$:<br>
$$\psi_{j+1} = (2 + \frac{2m}{\hbar^2}[V_j-E]\;a^2)\psi_j - \psi_{j-1}$$
<br>

This equation tells us that if we specify the values of $\psi_0$ and $\psi_1$, we can iteratively solve for the value of $\psi$ at all other points of our grid. 



### To the left of the barrier, at $x=0$, the wave is equal to $\psi(x=0)=c_2e^{-ik0} = c_2$, If we assume a unit amplitude of the transmitted wave, then $\psi(x=0)=c_2 = 1$.<br> At the next point in the grid, where $x=a$, we'll have $\psi(x=1)=c_2e^{-ika}$.  And those two points are enough to get our iterative scheme explained above going!

### We'll write a cell of code for you that creates a potential barrier of a give width, height, and center position:

```python
#Set the potential barrier:
def V_barrier(x_array, width = 1, V0=9, x0 = 4):
    Vx = np.zeros_like(x_array)
    for i in range(len(x_array)): 
        if  x_array[i]<x0:
            Vx[i] = 0
        elif x_array[i] < x0+width:
            Vx[i] = V0
        else:
            Vx[i] = 0
    return Vx
            
```

### Create a grid of x-coordinates called xcoords, from 0 to 10 in intervals of 0.01.  Then plot the potential $V(x)$ in that range of x values. 

```python
#
#grid of x coordinates
a = 0.01
xmax = 10
xcoords = np.arange(0, xmax, a)

###########################
#get Vx, the barrier potential for that range of values, and 
#plot it as a function of x. 
Vx = V_barrier(xcoords)
#########################
plt.plot(xcoords, Vx, linewidth=2)
plt.xlabel('x')
plt.ylabel('V(x)')
plt.ylim(0,10)
```

### Now let's define some of the relevant constants

```python
#atomic units
hbar = 1
m = 1
#energy and wavenumber
energy = 9
k = np.sqrt(2*m*energy/hbar**2)
```

### Next, we'll build our wave function iteratively using the technique described above. To do that, you will first need to create an array of zeros, called psi, of the same length as xcoords. And we'll assign the values of $\psi[0]$ and $\psi[1]$ ourselves. 

```python
#wavefunction array , zeros
psi = np.zeros(len(xcoords), dtype=np.dtype(complex))

####################################################################################
#Two lines Assign values of wave function: psi[0] and psi[1]
psi[0] = 0
psi[1] = exp(-1j*k*a)
#################################################################################
```

### Next, write a for loop that goes from $i=1$ to $i = len(xcoords)-1$. And for each value of $i$, get the value of $\psi[i+1]$ using the finite differences approach outlined above. After running the for loop, plot the real part of $\psi(x)$ as a function of coordinates.   $\psi_{j+1} = (2 + \frac{2m}{\hbar^2}[V_j-E]\;a^2)\psi_j - \psi_{j-1}$

```python
#iterate to get all values of psi
for i in range(1, len(xcoords)-1):   
    ####################################################################
    #One line, assign the value of psi[i+1] using the 
    #finite differences relation outlined above. 
    ####################################################################
    psi[i+1] = (2+2*m/hbar**2*(Vx[i]-energy)*a**2)*psi[i]-psi[i-1]

#Plot the real part of psi, overlayed with the potential: 
psi.real = psi.real * .05
quantumWorld.chem160_plotting([xcoords, xcoords], [psi.real, Vx], 'TunnelingThroughBarrier',
                              ('psi','V(x)'), 'x', ylabel='V(x)', ylimits=(-4,10))    
plt.plot(xcoords, psi.real)
plt.plot(xcoords, Vx)
```

## Computing the transmission probability

### The transmission probability is the ratio between the "amount" of particles that made it through the barrier to the "amount" of particles that was sent in. We use probability densities for "amounts": $$t = \|\frac{c_2}{c_0}\|^2$$  **Following Schrier, section 3.1.3, one can derive the following relation: ** $$ t = \frac{2}{1 + P_{avg}} $$ <br> Where $P_{avg}$ can be obtained by taking the average of the minimum and maximum values of the oscillations of the PDF associated to the wave packet sufficiently far away to the right of the barrier. 

### Write a function called transmission_probability that takes as input the pdf associated to a wavefunction and returns the transmission probability

```python
def transmission_probability(pdf):
    Pmin = np.min(pdf[-300:])
    Pmax = np.max(pdf[-300:])
    
    ########################
    #In two lines, get 1) Pavg (P average) which is (Pmin+Pmax)/2
    #and 2) the transmission probability, (call it transmission)
    Pavg = (Pmin+Pmax)/2.0
    transmission = 2.0/(1+Pavg)
    ##########################
    
    return transmission
```

## Next write a function called transmission_through_barrier that takes as input the xcoords array and a set of parameters (mass of particle, energy, barrier width), and returns that transmission probability. 

```python
def transmission_through_barrier(xcoords, m = 1, energy = 9, width = 1):
   ################################################3
    #Get potential by calling V_barrier
    ####################################################
    Vx = V_barrier(xcoords, width=width)
    ################################################################
    #get the wave number k, for the given energy: 
    ##############################################################
    k = np.sqrt(2*m*energy/hbar**2)
    #Build wave function: 
    #wavefunction array , zeros
    psi = np.zeros(len(xcoords), dtype=np.dtype(complex))
    
    #Assign values of wave function at j = 0 and j=1
    psi[0] = 1
    psi[1] = np.exp(-1j*k*a)
    
    ###########################################################################
    #iterate to get all values of psi
    for i in range(1, len(xcoords)-1):   
        #One line, assign the value of psi[i+1] using the 
        #finite differences relation outlined above.
        psi[i+1] = (2+2*m/hbar**2*(Vx[i]-energy)*a**2)*psi[i]-psi[i-1]
    #######################################################################
    
    #####################################################################
    #Get pdf associated to psi
    ############################################################################
    pdf = quantumWorld.probabilityDensity(psi)
    ####################################3
    #Get transmission probabilty by calling the function above, 
    #transmission_probability
    transmission = transmission_probability(pdf)
    #################################################################
    
    return transmission
```

### Run the function above for a range of energy values that goes from $E=0$ to $E=25$ in steps of 0.1. Store the value of the transmission probabilities in an array.  And Plot transmission probability   

```python
#grid
a = 0.01
xTotal = 10
nPoints = xTotal/a
xcoords = np.arange(0, xTotal, a)
m=1

##############################################################
#Create energy_array 
energy_array = np.arange(0, 25, 0.1)
#######################################

transmission_list = []

################################################
for energy in energy_array: 
    #####################
    #assign to variable transmission the transmission prob at the value 
    #energy 
    transmission = transmission_through_barrier(xcoords, m, energy)
    ###################
    
    transmission_list.append(transmission)
############################################################
```

### Plot transmission as a function of energy:

```python
quantumWorld.chem160_plotting([energy_array], [transmission_list], 'Transmission vs. Energy', None, 'Energy', 'Transmission')
```

```python
#grid
a = 0.01
xTotal = 10
nPoints = xTotal/a
xcoords = np.arange(0, xTotal, a)

energy = 9 

#############################################
#in about 5 lines, create a mass_array, an empty transmission_list, 
#and write a for loop that gets the transmission probability for each 
#value of mass, and stores it in transmission_list. 
##############################################
mass_array = np.arange(0.5, 5, 0.1)
transmission_list = []

for m in mass_array: 
    transmission = transmission_through_barrier(xcoords, m, energy)    
    transmission_list.append(transmission)
```
